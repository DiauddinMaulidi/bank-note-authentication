```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler 
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import cross_val_score
```

```python
# membaca dataset
data = pd.read_csv('data_banknote_authentication.csv', header=None)
data.columns = ['variance', 'skewness', 'kurtosis', 'entropy', 'class']
data
```

#### menganalisa dan menyiapkan data

```python
# cek data kosong
data.info()
```

```python
# deskripsi dataset
data.describe()
```

```python
# jumlah pengamatan setiap kelas
data['class'].value_counts()
```

```python
# visualisasi jumlah pengamatan setiap kelas
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='class', hue='class')
plt.title('Distribusi Kelas (0 = Asli, 1 = Palsu)')
plt.xlabel('Class')
plt.ylabel('Jumlah')
plt.tight_layout()
plt.show()
```

#### Preprocessing

```python
# memisahkan fitur dan targetnya
X = data.drop('class', axis=1)
y = data['class']
```

```python
# menormalkan data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
print(X_scaled)
```

```python
# mereduksi dimensi
pca = PCA(n_components=2)
dataPCA = pca.fit_transform(X_scaled)
```

```python
print(dataPCA)
```

```python
# menjadikan fitur baru hasil setelah PCA menjadi dataframe
principalDf = pd.DataFrame(data = dataPCA, columns = ['PC1', 'PC2'])
# menggabungkan fitur baru dengan data target atau kelas
finalDf = pd.concat([principalDf, y], axis = 1)
finalDf
```

```python
# visualisasi hasil PCA
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_title('Visualisasi PCA')

targets = [0, 1]
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['class'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'PC1'], finalDf.loc[indicesToKeep, 'PC2'], c = color, s = 15)
ax.legend(targets)
ax.grid()
```

```python
# hubungan setiap fitur
plt.figure(figsize=(8,6))
sns.heatmap(X.corr(), annot=True, cmap='plasma')
plt.title('Matriks Korelasi')
plt.show()
```

```python
# memisahkan data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, train_size=0.8)
```

```python
# smt = SMOTE()
# X_train_sm, y_train_sm = smt.fit_resample(X_train, y_train)
```

#### Pemodelan

```python
algoritma = ['Decision Tree', 'K-Nearest Neighbors', 'Naive Bayes', 'SVM']
models = [DecisionTreeClassifier(), KNeighborsClassifier(n_neighbors=7), GaussianNB(), SVC()]

for model, name in zip(models, algoritma):
    # print(model)
    model.fit(X_train, y_train)

    train_pred = model.predict(X_train)
    train_acc = accuracy_score(y_train, train_pred)
    print(f"Akurasi di data training: {train_acc:.2f}")
    
    predict_test = model.predict(X_test)
    print(f"Accuracy {name}: ", accuracy_score(y_test, predict_test))
    print(f"{classification_report(y_test, predict_test)}\n")

    scores = cross_val_score(model, X_scaled, y, cv=10)
    print(f"Akurasi CV: {scores}")
    
    fig, ax = plt.subplots(figsize=(8,5))
    sns.heatmap(confusion_matrix(y_test, predict_test), annot=True, fmt=".0f", ax=ax)
    plt.xlabel("Predicted label")
    plt.ylabel("True Label")
    plt.show()
```
